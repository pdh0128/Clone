```python
import requests
from bs4 import BeautifulSoup

r = requests.get("http://www.dowellcomputer.com/main.jsp")
html = r.text
soup = BeautifulSoup(html, 'html.parser')
links = soup.select('td > a')

for link in links:
    if link.has_attr('href'):
        if link.get('href').find('study') != -1:
            print(link.text) 
```
